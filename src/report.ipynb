
## Overview

These notes summarize the replication of **Figure 1** and **Tables 1, 3, and 4**.


## Data

Monthly with and without dividend returns of the value weighted portfolio of all NYSE, Amex, and NASDAQ stocks from 1944 to 2007 and 2024.

30-day T-bill rate from 1944 to 2007 and 2024.

The reason for needing the data from 1944 instead of 1946 was due to the construction of certain parameters requiring data from previous years. 


## Notation

For month $m$,

- $r_m^d$ = monthly value-weighted return **with** dividends (`vwretd`),
- $r_m^x$ = monthly value-weighted return **without** dividends (`vwretx`),
- $r_m^f$ = monthly T-bill return (`t30ret`),
- $P_m$ = ex-dividend price index at month $m$,
- $\mathrm{Div}_m$ = dividend cash flow paid during month $m$.



## Monthly-to-Annual Construction

### 1. Ex-dividend price index

The ex-dividend return satisfies

$$
1+r_m^x = \frac{P_m}{P_{m-1}}.
$$

So the ex-dividend price index is built recursively as

$$
P_m = P_{m-1}(1+r_m^x),
$$

starting from an arbitrary normalization such as $P_0=1$.  

### 2. Monthly dividend cash flow

The cum-dividend return satisfies

$$
1+r_m^d = \frac{P_m+\mathrm{Div}_m}{P_{m-1}}.
$$

Subtracting the ex-dividend equation gives

$$
\mathrm{Div}_m = P_{m-1}\bigl(r_m^d-r_m^x\bigr).
$$

This is the monthly dividend cash flow.

### 3. Annual dividends under two reinvestment conventions

For each calendar year $t$, the monthly dividends are carried to the end of that year in two different ways.

#### Risk-free reinvestment

$$
D_t
=
\sum_{m \in t}
\mathrm{Div}_m
\prod_{\substack{j>m \\ j \in t}}
(1+r_j^f).
$$

#### Market reinvestment

$$
D_t^M
=
\sum_{m \in t}
\mathrm{Div}_m
\prod_{\substack{j>m \\ j \in t}}
(1+r_j^d).
$$

### 4. Annual dividend growth

The annual log dividend-growth series are

$$
\Delta d_t
=
\log\left(\frac{D_t}{D_{t-1}}\right),
\qquad
\Delta d_t^M
=
\log\left(\frac{D_t^M}{D_{t-1}^M}\right).
$$

These two series are the objects behind both **Figure 1** and **Table 1**.

### 5. Annual returns and log price-dividend ratios

Let $P_t$ denote the end-of-year ex-dividend price index. Then

$$
r_t
=
\log\left(\frac{P_t+D_t}{P_{t-1}}\right),
\qquad
r_t^M
=
\log\left(\frac{P_t+D_t^M}{P_{t-1}}\right).
$$

The annual log price-dividend ratios are

$$
pd_t = \log\left(\frac{P_t}{D_t}\right),
\qquad
pd_t^M = \log\left(\frac{P_t}{D_t^M}\right).
$$

## Table 4

Table 4 studies reduced-form time-series models for market-reinvested dividend growth. Let

$$
y_t := \Delta d_t^M.
$$

### Column 1: AR(1)

$$
y_t = a_1 + b_1 y_{t-1} + u_t.
$$

This was estimated by OLS.

### Column 2: lagged-return regression

$$
y_t = a_2 + b_2 r_{t-1} + u_t.
$$

This was also estimated by OLS.

### Column 3: direct ARMAX(1,1)

The final specification was estimated as

$$
y_t = c + \phi y_{t-1} + \beta r_{t-1} + e_t + \theta e_{t-1},
$$

with $e_0=0$.

Rearranging gives the innovation recursion

$$
e_t = y_t - c - \phi y_{t-1} - \beta r_{t-1} - \theta e_{t-1}.
$$

For a candidate parameter vector

$$
\psi = (c,\phi,\theta,\beta),
$$

the conditional innovation variance is

$$
\hat{\sigma}^2(\psi) = \frac{1}{T}\sum_{t=1}^T e_t(\psi)^2.
$$

The conditional Gaussian negative log-likelihood minimized in the replication is

$$
Q(\psi)
=
\frac{T}{2}
\left[
\log(2\pi)+1+\log \hat{\sigma}^2(\psi)
\right].
$$

### Reported constant in column 3

The raw intercept is $c$, but the reported column-3 constant was taken to be the unconditional mean

$$
\gamma_0
=
\frac{c+\beta \bar r}{1-\phi},
\qquad
\bar r = \frac{1}{T}\sum_{t=1}^T r_{t-1}.
$$

Its standard error was computed using the delta method with a covariance matrix based on the numerical Hessian of the objective.

Replication was successful.

## The Challenge

The main difficulty occurred in **Table 4, column 3**.

### Initial ARIMA attempt

The first attempt used a generic ARIMA model with an exogenous regressor. Conceptually, that looks like

$$
y_t = \alpha + \beta r_{t-1} + u_t,
$$

where the disturbance $u_t$ follows an ARMA(1,1) process.

So the ARMA structure is placed on the **regression error**, not directly on the observable recursion used later.

Under that approach, the resulting column-3 estimates were unsatisfactory.

### Direct ARMAX replacement

To fix this, the ARIMA-with-exogenous-regressor approach was replaced by **direct estimation** of

$$
y_t = c + \phi y_{t-1} + \beta r_{t-1} + e_t + \theta e_{t-1}.
$$

The main changes were:

1. the model was rewritten in **innovation recursion form**,
2. the optimizer no longer estimated a generic `ARIMA(..., exog=...)` object,
3. instead, it directly minimized the conditional Gaussian likelihood based on the recursively defined innovations,
4. several starting values were tried for $(c,\phi,\theta,\beta)$,
5. $\phi$ and $\theta$ were constrained numerically away from $\pm 1$,
6. the reported constant was changed from a raw intercept-type quantity to the unconditional mean
   $$
   \gamma_0 = \frac{c+\beta \bar r}{1-\phi},
   $$
7. its standard error was computed using the numerical Hessian and the delta method.

### Why the direct ARMAX version was better

After this change, the estimates were much better.

The key issue was that the original ARIMA-with-exog formulation was not matching the direct reduced-form recursion that the final replication intended to estimate.



## Conclusion

The overall replication was successful. The small differences were likely due to the paper using data from CRSP while us using data from WRDS.
